### **Qual a Receita Total?**


```
dados["Receita"] = dados["Valor Venda"].mul(dados["Quantidade"])
dados.head()
```

### **Qual o custo Total?**


```
dados["Receita-Custo Unitário"] = dados["Receita"] - dados["Custo Unitário"]
```

```
dados.head()
```

### **Qual o Lucro Total?**


```
dados.groupby("ID Loja")["Receita"].sum()
```

### **Qual foi o Lucro por ano?**


```
dados.groupby(dados['Data Venda'].dt.year)['Receita-Custo Unitário'].sum()
```

### **Qual o total de produtos vendidos?**


```
dados.groupby(dados["Quantidade"]).sum()
```
